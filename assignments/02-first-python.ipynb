7. To calculate Fibonacci Series up to n numbers.


```python
def fibonacci(n):    
    a,b = 1,1
    print(a,b, sep=", ", end=", ")

    for _ in range(3, n+1):
        c = a+b
        print(c, end=", ")
        a, b = b, c

fibonacci(10)
```

8. To find out whether the given String is Palindrome or not.


```python
def palindrome(string):
    return string == string[::-1]

print(palindrome("ankur"))
print(palindrome("nayan"))
```

9. To find Armstrong Number between two given number.

An Armstrong number is a number that is equal to the sum of its own digits each raised to the power of the number of digits in the number.



```python
def is_armstrong(n):
    digits = []
    tmp = n
    while tmp != 0:
        remainder = tmp % 10
        digits.append(remainder)
        tmp = tmp // 10
    total = sum([i**len(digits) for i in digits])

    return total == n

def find_armstrong(n,m):
    for i in range(n, m+1):
        if is_armstrong(i):
            print(i)


find_armstrong(100, 500)
```
