1. [Define a method to find out if a number is prime or not.](https://www.geeksforgeeks.org/java-program-to-check-if-a-number-is-prime-or-not/)


```python
def is_prime_number(n):
    if n == 1 or n == 2:
        return True
    
    i = 2
    while (i <= n/2 + 1):
        if n % i == 0:
            return False
        i += 1
    
    return True

print(is_prime_number(2))
print(is_prime_number(3))
print(is_prime_number(5))
print(is_prime_number(7))
print(is_prime_number(8))
print(is_prime_number(9))
```


5. [Write a program to print the factorial of a number by defining a method named 'Factorial'.](https://www.javatpoint.com/factorial-program-in-java)
Factorial of any number n is represented by n! and is equal to 1 * 2 * 3 * .... * (n-1) *n. E.g.- <br/>
<pre>
4! = 1 * 2 * 3 * 4 = 24 
3! = 3 * 2 * 1 = 6 
2! = 2 * 1 = 2 
Also, 
1! = 1 
0! = 1
</pre>



```python
def factorial(n):
    product = 1
    while n != 0:
        product = product * n
        n = n - 1
    return product

print(factorial(5))
```

6. [Write a function to find if a number is a palindrome or not. Take number as parameter.](https://www.geeksforgeeks.org/check-if-a-number-is-palindrome/)




```python
def num_palindrome(n):
    num_str = str(n)
    return num_str == num_str[::-1]

print(num_palindrome(121))
print(num_palindrome(12))
print(num_palindrome(123))
```

7. [Write a function to check if a given triplet is a Pythagorean triplet or not.](https://www.geeksforgeeks.org/find-pythagorean-triplet-in-an-unsorted-array/) (A Pythagorean triplet is when the sum of the square of two numbers is equal to the square of the third number).


```python
def is_pythagorean_triplet(a,b,c):
    a,b,c = sorted([a,b,c])
    return (a*a + b*b) == c*c

print(is_pythagorean_triplet(5,4,3))
print(is_pythagorean_triplet(1,4,3))
```


8. [Write a function that returns all prime numbers between two given numbers.](https://www.geeksforgeeks.org/program-to-find-prime-numbers-between-given-interval/)


```python
def prime_number_in_range(n,m):
    for i in range(n,m+1):
        if is_prime_number(i):
            print(i)

prime_number_in_range(1, 20)
```
